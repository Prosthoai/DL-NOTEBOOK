#**DATA preprocessing**

```python
import numpy as np 
from random import randint 
from sklearn.preprocessing import MinMaxScaler 
```

```python
train_labels=[] 
train_samples=[]
```

Example data: 



*   An experimental drug was tested on individuals from ages 13 to 100
*   The trial had 2100 participants. Half were under 65 yrs old., and half were over 65 yrs old.
*   95% of patients 65 or older experienced side effects
*   95& of patients under 65 experienced no side effects






```python
for i in range (1000):
  random_younger=randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)

  random_older=randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)

for i in range (50):
  random_younger=randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1)

  random_older=randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0)

```

```python
#print raw data
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
  print(i)
```

```python
train_labels=np.array (train_labels)
train_samples=np.array(train_samples)
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
scaled_train_samples=scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
#print scaled data
for i in scaled_train_samples:
  print(i)
```

transform the data into between 0 and 1, so that the net can make more accurate predictions on the data
